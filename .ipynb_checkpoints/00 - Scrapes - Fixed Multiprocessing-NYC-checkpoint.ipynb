```python
import datetime

# Scraping items
from bs4 import BeautifulSoup
import requests

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

import os
chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
#chrome_options.add_argument('--no-sandbox')
#chrome_options.add_argument('--disable-dev-shm-usage')

from IPython.core.display import display, HTML

from datetime import datetime as dt

from pymongo import MongoClient

client = MongoClient()
db = client.fletcher
```

```python
import concurrent.futures
```

```python
#driver = webdriver.Chrome(chromedriver) 

def get_page_url(base_url, page=0):
    url_parts = base_url.split('Reviews')
    new_base = url_parts[0] + 'Reviews'
    if page == 0:
        return base_url
    return new_base + f'-or{page*10}'

def get_expanded_html(driver, base_url, page=0):
    url = get_page_url(base_url, page)
    driver.get(url)
    time.sleep(5)
    
    try:
        locate = driver.find_element_by_xpath('.//span[@class = "taLnk ulBlueLinks"]')
        time.sleep(1)
        locate.click()
        time.sleep(3)
    except:
        print("This is not working yet")
        
    return driver.page_source

def put_into_mongo(documents, db=db):
    db.wash_revs.insert_many(documents)

def scrape_html(source):
    soup = BeautifulSoup(source, 'html.parser')
    
    site_name = soup.find(id='HEADING').text
    print(site_name)
    
    all_reviews_selectors = ["rev_wrap ui_columns is-multiline", 
                             "review hsx_review is-multiline is-mobile inlineReviewUpdate provider0"]
    
    all_reviews = []
    for selector in all_reviews_selectors:
        all_reviews = soup.find_all(class_=selector)
        if len(all_reviews)>0:
            break
                
    documents = []
    for review in all_reviews:
        # review title
        
        # review text
        review_text = [review.find(class_="partial_entry").get_text().replace('\n', '')]

        # associated rating
        rating = review.find('span', class_='ui_bubble_rating')['class'][1].split('_')[1]
        # associated author

        try:
            author = review.find(class_='info_text').find('div').get_text()
        except:
            author = review.find(class_='expand_inline scrname').get_text()
        
        # author location

        try:
            location = review.find(class_='info_text').find(class_='userLoc').get_text()
        except:
            try:
                location = review.find(class_='expand_inline userLocation').get_text()
            except:
                location = ''
        
        # date review was written

        date = review.find(class_='ratingDate')['title']
        try:
            date = datetime.datetime.strptime(date, '%B %d, %Y')
        except ValueError:
            pass
        
        document = {
            'text': review_text,
            'date': date,
            'location': location,
            'rating': int(rating),
            'author': author,
            'site': site_name
        }
        documents.append(document)
    
    put_into_mongo(documents)
    
    try:
        current = int(soup.select('a.pageNum.current')[0].text)
    except AttributeError:
        current = 0
    
    try:
        last = int(soup.select('a.pageNum.last.taLnk')[0].text)
    except AttributeError:
        last = 0
    print(current, last)    
    return (current, last)
```

```python
def scrape_all(base_url, starting_page=0):
    
    driver = webdriver.Chrome(chromedriver) 
    current = starting_page
    last = float('inf')
    failed_attempts = 0
    
    while current < last:
        src = get_expanded_html(driver, base_url, current)
        try:
            current, last = scrape_html(src)
            failed_attempts = 0
        except:
            time.sleep(0.5)
            failed_attempts += 1
            
        if failed_attempts > 3:
            print(f"Scraped {current} pages out of {last}, incurred too many consecutive failures")
            return False
        
        if (current % 5 == 0):
            print(f"Finished page {current} out of {last}")
            
    return 'Done'
```

## DO NOT DELETE

```python
counts = []
with concurrent.futures.ProcessPoolExecutor() as executor:
    well = [executor.submit(scrape_all, url) for url in nyc]
    counts.append(well)
```
